```python
import pandas as pd       
import numpy as np        
import glob               
import os                
import warnings
warnings.filterwarnings('ignore')  

print("Libraries imported successfully!")
print(f"   pandas version : {pd.__version__}")
print(f"   numpy  version : {np.__version__}")
```

```python
csv_files = glob.glob('*.csv')
xls_files = glob.glob('*.xls') + glob.glob('*.xlsx')
all_files  = csv_files + xls_files

print(f"FILES FOUND:")
print(f"   CSV files  : {len(csv_files)}")
print(f"   XLS files  : {len(xls_files)}")
print(f"   Total files: {len(all_files)}")

if len(all_files) == 0:
    print("WARNING: No files found! Make sure your CSV/XLS files are in the same folder as this notebook.")
else:
    print(f"   First 5 files: {all_files[:5]}")
```

```python
all_dataframes = []    
failed_files   = []    
success_count  = 0

print("Reading all files...")
print("-" * 55)

for file in all_files:
    try:
        # Read CSV files
        if file.endswith('.csv'):
            df_temp = pd.read_csv(file, encoding='latin-1')
        # Read XLS/XLSX files
        else:
            df_temp = pd.read_excel(file)

        # Add a column showing which file this row came from
        # os.path.splitext removes extension: 'backpacks.csv' -> 'backpacks'
        df_temp['source_file'] = os.path.splitext(file)[0]

        all_dataframes.append(df_temp)
        success_count += 1
        print(f"   OK  {file:40} -> {len(df_temp):>4} rows")

    except Exception as e:
        failed_files.append(file)
        print(f"   FAILED  {file} -> {e}")

print("-" * 55)
print(f"Successfully loaded : {success_count} files")
if failed_files:
    print(f"Failed to load     : {len(failed_files)} files -> {failed_files}")
```

```python
df = pd.concat(all_dataframes, ignore_index=True, sort=False)

print(f"ALL FILES MERGED SUCCESSFULLY!")
print(f"   Total rows    : {len(df):,}")
print(f"   Total columns : {df.shape[1]}")
print()
print("Preview of merged DataFrame:")
df.head()
```

```python
missing_counts  = df.isnull().sum()
missing_percent = (df.isnull().sum() / len(df) * 100).round(2)

missing_summary = pd.DataFrame({
    'Missing Count' : missing_counts,
    'Missing %'     : missing_percent
})

missing_summary = missing_summary[missing_summary['Missing Count'] > 0].sort_values(
    'Missing Count', ascending=False
)

print(f"MISSING VALUES REPORT:")
print(f"   Total missing cells  : {df.isnull().sum().sum():,}")
print(f"   Columns with missing : {len(missing_summary)}")
print("-" * 50)
print(missing_summary)
```

```python
original_rows    = len(df)
original_missing = df.isnull().sum().sum()


df_clean = df.copy()

numeric_cols = df_clean.select_dtypes(include=[np.number]).columns

for col in numeric_cols:
    missing_in_col = df_clean[col].isnull().sum()
    if missing_in_col > 0:
        median_val = df_clean[col].median()
        df_clean[col] = df_clean[col].fillna(median_val)
        print(f"   Numeric '{col}': filled {missing_in_col} missing -> median = {median_val:.2f}")

text_cols = df_clean.select_dtypes(include=['object']).columns

for col in text_cols:
    missing_in_col = df_clean[col].isnull().sum()
    if missing_in_col > 0:
        df_clean[col] = df_clean[col].fillna('Unknown')
        print(f"   Text    '{col}': filled {missing_in_col} missing -> 'Unknown'")

threshold        = int(df_clean.shape[1] * 0.5)
rows_before_drop = len(df_clean)
df_clean         = df_clean.dropna(thresh=threshold)
rows_dropped_missing = rows_before_drop - len(df_clean)

print()
print(f"Missing values handled!")
print(f"   Remaining missing cells : {df_clean.isnull().sum().sum()}")
print(f"   Rows dropped            : {rows_dropped_missing}")
```

```python
useful_cols = [
    'product_id', 'title', 'category',
    'initial_price', 'discount', 'final_price',
    'rating', 'ratings_count', 'sizes', 'currency'
]


useful_cols = [col for col in useful_cols if col in df_clean.columns]

df_selected = df_clean[useful_cols]

print(f"SELECTED {len(useful_cols)} KEY COLUMNS:")
print(f"   Columns: {useful_cols}")
print("-" * 60)
df_selected.head(10)
```

```python
num_duplicates = df_clean.duplicated().sum()

print(f"DUPLICATE CHECK:")
print(f"   Total rows           : {len(df_clean):,}")
print(f"   Duplicate rows found : {num_duplicates:,}")

if num_duplicates > 0:
    print("\n   Sample duplicate rows:")
    print(df_clean[df_clean.duplicated()][['product_id', 'title', 'category']].head(3))
else:
    print("   No duplicates found - dataset is clean!")
```

```python
rows_before = len(df_clean)

df_clean = df_clean.drop_duplicates(keep='first')

df_clean = df_clean.reset_index(drop=True)

duplicates_removed = rows_before - len(df_clean)

print(f"Duplicates Removed!")
print(f"   Rows before  : {rows_before:,}")
print(f"   Rows removed : {duplicates_removed:,}")
print(f"   Rows after   : {len(df_clean):,}")
```

```python
# Create total_amount = initial_price x ratings_count
df_clean['total_amount'] = df_clean['initial_price'] * df_clean['ratings_count']

print("Created 'total_amount' column!")
print("   Formula : total_amount = initial_price x ratings_count")
print("   Meaning : Estimated total revenue per product")
print()

print("New Column Preview:")
print("-" * 60)
print(
    df_clean[['title', 'category', 'initial_price', 'ratings_count', 'total_amount']]
    .head(10)
    .to_string(index=False)
)

print()
print("total_amount Statistics:")
print(f"   Min    : {df_clean['total_amount'].min():>15,.2f}")
print(f"   Max    : {df_clean['total_amount'].max():>15,.2f}")
print(f"   Mean   : {df_clean['total_amount'].mean():>15,.2f}")
print(f"   Median : {df_clean['total_amount'].median():>15,.2f}")
print(f"   Total  : {df_clean['total_amount'].sum():>15,.2f}")
```
